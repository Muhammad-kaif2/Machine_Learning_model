```
# Ridge Regression Linear Regression ka ek variant hai jisme L2 regularization add ki jati hai.
Ye model ke coefficients (weights) ko shrink karta hai taake overfitting kam ho.
- Ye penalty model ke coefficients (βᵢ) ko chhote values tak compress / shrink kar deti hai.
Yani, large coefficients ko reduce kar diya jata hai taake model overfit na kare.
-Shrinking matlab 0 karna nahi, bas value ko smaller aur stable banana.
Iska effect: model generalize better karta hai, test data par overfitting kam hoti hai.

- from sklearn.linear_model import Ridge
```



````python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
````

````python
from sklearn.datasets import load_diabetes

data=load_diabetes()
````

````python

X=data.data
y=data.target
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
````

````python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
````

````python
L.fit(X_train,y_train)
````

````python
print(L.coef_)
print(L.intercept_)
````

````python
from sklearn.metrics import r2_score,mean_squared_error
acc_tra = r2_score(y_train,L.predict(X_train))
err = np.sqrt(mean_squared_error(y_train,L.predict(X_train)))
print(f'train_accurecy {acc_tra}')
print(f'train_err {err}')
````

````python
from sklearn.metrics import r2_score,mean_squared_error
acc_test = r2_score(y_test,L.predict(X_test))
err = np.sqrt(mean_squared_error(y_test,L.predict(X_test)))
print(f'test_accurecy {acc_test}')
print(f'test_err {err}')
````

````python
from sklearn.linear_model import RidgeCV
R=Ridge(alpha=0.1)
````

````python
R.fit(X_train,y_train)
````

````python
r_acc_tra = r2_score(y_train,R.predict(X_train))
r_err = np.sqrt(mean_squared_error(y_train,R.predict(X_train)))
print(f'train_accurecy {r_acc_tra}')
print(f'train_err {r_err}')
````

````python
r_acc_test = r2_score(y_test,R.predict(X_test))
r_err = np.sqrt(mean_squared_error(y_test,R.predict(X_test)))
print(f'test_accurecy {r_acc_test}')
print(f'test_err {r_err}')
````

**Ridge_reg- gradient-descent**

````python
X,y = load_diabetes(return_X_y=True)
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
````

````python
from sklearn.linear_model import SGDRegressor
````

````python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
````

````python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
````

````python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
````

````python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
````
